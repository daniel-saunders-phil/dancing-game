## Unpublished but helpful figures

```python
# depicts the frequencies of strategies when the model is configured to
# only contain two unconditional strategies
# always perform A
# always perform B
# this figure is not included in the published version of the paper

url = "https://raw.githubusercontent.com/daniel-saunders-phil/dancing-game/main/data/Typical%20unconditional%20strategy%20distribution.csv"
strategy_frequency = pd.read_csv(url,header=19)

x = strategy_frequency['x']
b_freq = strategy_frequency['y']
a_freq = strategy_frequency['y.1']

plt.plot(x, b_freq, label='always B')
plt.plot(x, a_freq, label='always A')
plt.xlabel('time')
plt.ylabel('strategy frequency')
plt.ylim([0,100])
plt.xlim([0,100])
plt.title('unconditional strategy frequencies')
plt.legend();
```

```python
# this figure depicts the average payoff for the condition described
# above
# this figure is not included in the published version of the paper

url = "https://raw.githubusercontent.com/daniel-saunders-phil/dancing-game/main/data/typical%20average-payoff%20for%202%20strategy%20uncoordinate.csv"

average_payoff = pd.read_csv(url,header=17)

x = average_payoff['x']
y = average_payoff['y']

middle = [.5] * 101

plt.plot(x, y)
plt.plot(x, middle, '--', color='gray', label='0.5')
plt.xlabel('time')
plt.ylabel('average payoff')
plt.ylim([0,1])
plt.xlim([0,100])
plt.legend()
plt.title('payoffs with unconditonal strategies');
```

```python
# this figure depicts the average payoff of the population
# for the condition where they employ gendered social learning
# and have the full range of strategies
# this figure is not included in the published version of the paper

url = "https://raw.githubusercontent.com/daniel-saunders-phil/dancing-game/main/data/typical%20average-payoff%20for%204%20strategy%20gendered%20learning.csv"

average_payoff = pd.read_csv(url,header=17)

x = average_payoff['x']
y = average_payoff['y']

middle = [.5] * 101
seventyfive = [.75] * 101

plt.plot(x, y)
plt.plot(x, middle, '--', color='gray', label='y = 0.5')
plt.plot(x, seventyfive, '--', color='gray', label='y = 0.75')
plt.xlabel('time')
plt.ylabel('average payoff')
plt.ylim([0,1])
plt.xlim([0,100])
plt.legend()
plt.title('payoffs with typical conditional strategies');
```

## Figure 2

```python
url = "https://raw.githubusercontent.com/daniel-saunders-phil/dancing-game/main/data/inertial%20mech%20strategy%20by%20sex.csv"

full_data = pd.read_csv(url, header=6)
full_data.head()

avg_payoff = full_data['global-average-payoff']

avg_po_as_str = str(round(avg_payoff.mean(), 3))

constant_avg_po = [avg_payoff.mean()] * len(avg_payoff)

plt.plot(avg_payoff,color='black')
plt.plot(constant_avg_po,"--",color="grey",label='Equilibrium Payoff')
plt.xlabel('Time')
plt.ylabel('Average Payoff')
plt.ylim([0,1])
plt.xlim([0,200])
plt.title('All Strategies & Random Selection of Learning Partners')
plt.legend()
plt.show();
```

## Table 2, 3, & 4

After publication, I found a more efficient way of representing the data all in one table. Table 2 in the paper is just the first row of table 3. Table 4 is just the second row.

```python
url = "https://raw.githubusercontent.com/daniel-saunders-phil/dancing-game/main/data/Endogenous%20gendered%20learning%20master.csv"

endo_experiment = pd.read_csv(url,header=4)

mr = endo_experiment['mutation_rate']
mls = endo_experiment['mutation_learning_styles']
success = endo_experiment['success']

mr_types = mr.value_counts().keys()
mr_types = list(mr_types)
mr_types.sort()
mls_types = mls.value_counts().keys()
mls_types = list(mls_types)
mls_types.sort()

raw_tab_data = []

for s in mr_types:
    
    row = [s]
    
    for ls in mls_types:
        s_group = (mr == s)
        ls_group = (mls == ls)
        success_w_s = success[s_group]
        success_target = success_w_s[ls_group]
        success_rate = round(success_target.mean() * 100,3)
        success_rate = str(success_rate) + '%'
        row.append(success_rate)
        
    raw_tab_data.append(row)

mls_types.insert(0, '')    
raw_tab_data.insert(0, mls_types)

try:
    from tabulate import tabulate
except ImportError:
    !pip install tabulate

print('rows = strategic mutation rate')
print('columns = learning style mutation rate')
print(tabulate(raw_tab_data, tablefmt='simple'))
```
